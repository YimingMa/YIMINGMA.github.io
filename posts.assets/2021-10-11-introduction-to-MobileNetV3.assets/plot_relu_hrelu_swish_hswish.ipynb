```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def swish(x):
    return x * sigmoid(x)
```

```python
def relu6(x):
    return np.array([min(max(0, ele), 6) for ele in x])
```

```python
def hsigmoid(x):
    return relu6(x + 3) / 6
```

```python
def hswish(x):
    return x * hsigmoid(x)
```

```python
x = np.arange(-6, 6.01, 0.01)
```

```python
plt.figure(figsize=(18, 6))
plt.plot(x, sigmoid(x), "--", label="sigmoid")
plt.plot(x, hsigmoid(x), label="h-sigmoid")
plt.legend(fontsize=20)
plt.grid(linestyle='-.')
plt.xlabel("x", fontsize=20)
plt.ylabel("f(x)", fontsize=20)
plt.axis([-6, 6, 0, 1.2])
plt.xticks(range(-6, 7, 2), fontsize=15)
plt.yticks(np.arange(0, 1.3, 0.2), fontsize=15)
plt.title("sigmoid vs. h-sigmoid", fontsize=25)
plt.savefig("sigmoid_vs_hsigmoid.png")
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.plot(x, swish(x), "--", label="swish")
plt.plot(x, hswish(x), label="h-swish")
plt.legend(fontsize=20)
plt.grid(linestyle='-.')
plt.xlabel("x", fontsize=20)
plt.ylabel("f(x)", fontsize=20)
plt.axis([-6, 6, -1, 8])
plt.xticks(range(-6, 7, 2), fontsize=15)
plt.yticks(np.arange(-1, 9, 1), fontsize=15)
plt.title("swish vs. h-swish", fontsize=25)
plt.savefig("swish_vs_hswish.png")
plt.show()
```
